# 介绍


1.数据集信息：
- 该数据集由属于 12,330 个会话的特征向量组成。
- 数据集的形成使得每个会话在一年内属于不同的用户，以避免任何特定活动的倾向，
- 特殊的日子、用户个人资料或时期。

2、属性信息：

该数据集由 10 个数字属性和 8 个分类属性组成。
- “收入”属性用作类标签，并且是我们预测模型中的目标特征。

- “管理”、“管理持续时间”、“信息”、“信息持续时间”、“产品相关”和“产品相关持续时间”表示访问者在该会话中访问的不同类型页面的数量以及在每个页面类别中花费的总时间。这些特征的值来源于用户访问的页面的URL信息并实时更新当用户采取行动时，例如从一页移动到另一页。

- “跳出率”、“退出率”和“页面价值”功能代表“Google Analytics”针对电子商务网站中每个页面测量的指标。网页的“跳出率”功能的值是指从该页面进入网站然后离开（“跳出”）的访问者的百分比在该会话期间不会触发对分析服务器的任何其他请求。特定网页的“退出率”功能的值是根据该页面的所有综合浏览量（会话中最后一次浏览的百分比）计算的。“页面价值”特征表示用户在完成电子商务交易之前访问的网页的平均价值。

- “特殊日期”功能表明网站访问时间与特定特殊日期的接近程度（例如母亲节、情人节）其中会话更有可能以交易结束。该属性的值是通过考虑电子商务的动态（例如订单日期和交货日期之间的持续时间）来确定的。例如，对于瓦伦蒂娜节，该值在 2 月 2 日到 2 月 12 日之间取非零值，在此日期之前和之后为零，除非它接近另一个特殊日子，并且在 2 月 8 日其最大值为 1。

- 数据集还包括操作系统、浏览器、区域、流量类型、访客类型（回访者或新访客）、指示访问日期是否为周末的布尔值以及一年中的月份。

现在我们将处理异常值。首先，我们要评估删除异常值对目标特征值数量的影响。


```python
#创建一个原始的数据副本

df_copy1 = df.copy()

#Create a dataframe with aggregation calculation: quantile 25th and 75th.
stats = df_copy1[['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues']].describe().transpose()

# stats = df_copy1[['Administrative_Duration','Informational_Duration','ProductRelated_Duration']].describe().transpose()

#创建一个新列"iqr"
stats['iqr'] = stats['75%']-stats['25%']

stats
```

```python
import pandas as pd

outliers = pd.DataFrame(columns=['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues'])

# outliers = pd.DataFrame(columns=['Administrative_Duration','Informational_Duration','ProductRelated_Duration'])
 
for col in stats.index:
    iqr = stats.at[col,'iqr']
    cutoff = iqr * 1.5
#     cutoff = iqr*1.7
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = df_copy1[(df_copy1[col] < lower) | 
                   (df_copy1[col] > upper)].copy()
    results['Outlier'] = col
outliers = pd.concat([outliers, results], ignore_index=True)

   
```

```python
outliers.index
```

```python
#Create a list of index of the outliers so that we can use to drop directly the corresponding rows.
lst_outliers = outliers.index.tolist()

print(f'The number of outliers is:',len(lst_outliers))
```

```python
# 确保lst_outliers中的索引存在于df_copy1中
valid_indices = df_copy1.index.intersection(lst_outliers)
if len(valid_indices) != len(lst_outliers):
    print("警告: 不是所有指定的索引都在df_copy1中，将仅删除存在的索引对应的行。")
    lst_outliers = valid_indices.tolist()  # 更新lst_outliers为有效索引

# 根据有效的索引列表删除行
df_copy1_no_out = df_copy1.drop(lst_outliers).reset_index(drop=True)
```

### Decision Tree

***创建函数***

```python
# 定义一个函数来确定决策树模型最相关的深度级别

def dt_mean_score(x_train, y_train, x_test, y_test):
    # 我们可以选择决策树的深度来决定模型细节程度，首先尝试深度值从1到25，并观察
    # 在给定的测试数据和标签上的平均准确率变化
    
    # 初始化一个空列表用于存储不同深度时的得分
    scores = []
    # 遍历1到25（包含25），尝试不同的深度值
    for n in range(1, 26):
        # 创建一个决策树分类器，并设置最大深度为当前值n
        tree = DecisionTreeClassifier(max_depth=n)
        # 使用训练数据拟合模型
        tree.fit(x_train, y_train)
        # 计算并在测试集上评估模型的准确率，然后将结果添加到scores列表
        scores.append(tree.score(x_test, y_test))
    
    # 绘制不同深度对应的准确率得分曲线
    plt.plot(range(1, 26), scores)
    # 显示图表
    plt.show()
```

```python
# 定义决策树模型的函数

def decision_tree(x_train, y_train, x_test, y_test, depth_level):
    # 使用指定的最大深度（depth_level）创建决策树分类器，并根据训练数据进行拟合
    tree = DecisionTreeClassifier(max_depth=depth_level)
    tree.fit(x_train, y_train)
    
    # 利用训练好的模型对测试集进行预测
    y_pred_tree = tree.predict(x_test)
    
    # 计算模型质量指标
    # 混淆矩阵
    conf1 = confusion_matrix(y_test, y_pred_tree)
    # 准确率
    acc1 = accuracy_score(y_test, y_pred_tree)
    # 召回率
    rec1 = recall_score(y_test, y_pred_tree)
    # 精确率
    pre1 = precision_score(y_test, y_pred_tree)
    # F1分数
    f11 = f1_score(y_test, y_pred_tree)
    
    # 展示这些评估指标
    print("混淆矩阵:\n", conf1)
    print("准确率:", acc1)
    print("召回率:", rec1)
    print("精确率:", pre1)
    print("F1分数:", f11)
```

### KNeighbors

```python
# 定义 K-近邻 (KNN) 分类器的函数

def knn(x_train, y_train, x_test, y_test, k_number):
    # 创建 KNN 分类器实例，其中 k_number 是用户指定的邻居数量
    knn_model = KNeighborsClassifier(k_number)
    # 使用训练数据集对模型进行训练
    knn_model.fit(x_train, y_train)
    # 利用训练好的模型对测试集进行预测
    y_pred_knn = knn_model.predict(x_test)
    
    # 计算模型的评估指标
    conf_matrix = confusion_matrix(y_test, y_pred_knn)  # 混淆矩阵
    accuracy = accuracy_score(y_test, y_pred_knn)       # 准确率
    recall = recall_score(y_test, y_pred_knn)           # 召回率
    precision = precision_score(y_test, y_pred_knn)     # 精确率
    f1 = f1_score(y_test, y_pred_knn)                  # F1分数
    
    # 展示混淆矩阵及各评估指标的结果
    display(conf_matrix)
    print('准确率:', accuracy)
    print('召回率:', recall)
    print('精确率:', precision)
    print('F1分数:', f1)
```

```python
#为KNN模型找一个好的K

def find_k(x_train,y_train,x_test,y_test,k_max):
    #Compute error score for each k number and append them to a list
    error_rate = []
    for i in range(1,k_max):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train,y_train)
        pred_i = knn.predict(x_test)
        error_rate.append(np.mean(pred_i != y_test))
    
    #Plot error score
    plt.figure(figsize=(10,6))
    plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    print("最小损失:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))   
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def dbscan_clustering(X, eps=0.5, min_samples=5):
    """
    应用DBSCAN算法进行数据聚类，并评估聚类质量。
    
    参数:
    X: np.ndarray, 特征矩阵，形状为(n_samples, n_features)
    eps: float, DBSCAN算法中的eps参数，定义邻域距离阈值
    min_samples: int, DBSCAN算法中的min_samples参数，定义成为核心对象所需的邻域中的样本数
    
    返回:
    None, 直接显示聚类结果的轮廓系数和聚类分配结果
    """
    # 初始化DBSCAN模型
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    
    # 训练模型并进行预测（聚类）
    clusters = dbscan.fit_predict(X)
    
    # 计算并打印轮廓系数，评估聚类质量
    if len(np.unique(clusters)) > 1:  # 至少有两个簇才能计算轮廓系数
        score = silhouette_score(X, clusters)
        print(f"轮廓系数: {score}")
    else:
        print("只有一个簇或所有点都被标记为噪声，无法计算轮廓系数。")
    
    # 可视化聚类结果（假设数据是二维的，对于高维数据可视化需降维处理）
    if X.shape[1] == 2:
        sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=clusters, palette='viridis')
        plt.title('DBSCAN Clustering Result')
        plt.show()
    else:
        print("数据不是二维的，未进行可视化。")

# 示例使用
# 假设X_train是特征数据
# dbscan_clustering(X_train)
```

```python
from sklearn.datasets import make_moons  # 仅用于生成示例数据
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
def find_eps(X, min_samples=5, max_eps=5, step=0.1):
    """
    寻找DBSCAN的最佳eps值。
    
    参数:
    X: 特征矩阵，np.array
    min_samples: 成为核心对象所需的邻域中的最少样本数，默认为5
    max_eps: 探索的eps最大值，默认为5
    step: eps值的步长，默认为0.1
    
    返回:
    最佳eps值和对应的轮廓系数
    """
    eps_values = np.arange(0.1, max_eps, step)
    silhouette_scores = []

    for eps in eps_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        cluster_labels = dbscan.fit_predict(X)
        

        score = silhouette_score(X, cluster_labels)
        
        silhouette_scores.append(score)
    
    # 找到轮廓系数最大的eps值
    best_eps = eps_values[np.argmax(silhouette_scores)]
    best_score = max(silhouette_scores)
    
    # 绘制eps值与轮廓系数的关系图
    plt.figure(figsize=(10, 6))
    plt.plot(eps_values, silhouette_scores, marker='o')
    plt.title('Silhouette Score vs. Eps Value')
    plt.xlabel('Eps')
    plt.ylabel('Silhouette Score')
    plt.grid(True)
    plt.show()
    
    print(f"最佳eps值: {best_eps}, 对应的轮廓系数: {best_score}")
    return best_eps, best_score
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

## PCA

     主成分分析（PCA）理论上可以应用于连续特征和经过独热编码处理的类别特征。但是，PCA基于方差最大化原则来识别数据中的主要变化方向，这在本质上更适用于连续数值型数据。对于二元或类别数据，尤其是经过独热编码后，PCA可能无法有效地捕获特征间的实际关系，因为其尝试最小化的是平方偏差，这在逻辑上并不完全适合于非连续的属性。
     当数据集中同时包含连续变量和类别变量时，采用因子分析的混合数据方法（Factor Analysis of Mixed Data, FAMD）会是更为合适的选择。FAMD是一种能够处理混合类型数据（包括定量变量和定性变量）的降维技术，它通过计算不同类型的变量之间的相关性矩阵来工作，因此能够更好地理解和保留数据中的结构信息。
    首先应用PCA：尽管PCA在处理类别特征时可能有局限性，但通过实践可以直观地观察到其效果，作为一种基准比较。
    随后应用FAMD：对于包含连续和类别特征的数据集，FAMD能提供一个更加合理且信息丰富的降维结果，因为它直接考虑了不同变量类型间的内在联系。
    通过对比PCA和FAMD的降维结果，不仅可以深入理解每种方法在特定数据集上的适用性和局限性，还能依据模型的表现和业务需求选择最合适的降维策略，从而优化后续的数据分析或建模过程。这种实验性的比较方法有助于在实践中找到最适合项目的解决方案。

## 下采样和过采样

正如在前两次运行中观察到的，模型在预测目标特征的“真”值方面存在问题，表现为极低的召回率。这是因为数据集存在不平衡性：负面样本（False）远多于正面样本（True），比例大约为85:15。
为了解决这一不平衡问题，我们提出了两个可以在项目中实施的解决方案：
- 随机下采样多数类（Random Undersampling）50%：此方法通过随机选择并移除多数类中的样本，从而减少变换后数据中多数类样本的数量。这样可以使得两类样本数量更加接近，但可能会丢失多数类中的一些重要信息。
- 使用SMOTE（合成少数类过采样技术）对少数类进行过采样：SMOTE通过从少数类中随机选取一个样本，然后计算该样本的k个最近邻点，并在该样本与其邻居之间生成合成样本，以此来增加少数类的样本数量。这种方法可以有效丰富少数类的多样性，降低过拟合风险，因为生成的是新样本而非简单复制原有少数类样本。
- 我们未在本项目中采用**随机过采样（Random Oversampling）**技术，原因是该方法通过直接复制少数类样本会导致模型更容易过拟合，因为增加了数据集中已存在的模式的重复性，没有提供额外的信息或模式多样性。相比之下，SMOTE通过生成合成样本提供了更好的平衡策略，有助于模型更好地泛化到未见过的数据。
    

因此，在这部分，我们将执行选定的两种抽样技术——随机下采样多数类和使用SMOTE对少数类进行过采样，以解决数据不平衡问题。
具体步骤如下：
- 实施抽样：
        - 随机下采样：从占优势的类别中随机选择并移除一部分样本，比如移除50%的“False”样本，以减少多数类与少数类之间的差距。
        - SMOTE过采样：针对“True”类别的样本，运用SMOTE算法生成新的合成样本，增加少数类样本的数量，同时保持数据的多样性和避免过拟合。
- 特征选择：
在完成抽样操作，使数据集的类别分布更为均衡之后，我们将应用
**主成分分析（PCA）**技术来选择与目标变量最相关的特征。PCA能够将原始特征转换为一组互不相关的主成分，并通过选择最重要的几个主成分来降低数据的维度，同时尽可能保留数据的变异性。
通过这样的流程，我们预期能有效地处理数据不平衡问题，并通过PCA筛选出对模型预测最有贡献的特征，进而提升模型在处理此类不平衡数据时的性能，尤其是提高对“True”类别的识别能力（即召回率）。
